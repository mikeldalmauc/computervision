```python
n=20
l=256
im=np.zeros((l,l))
points = l * np.random.random((2,n**2))
im[(points[0]).astype(np.int),(points[1]).astype(np.int)]= 1
im= filters.gaussian(im,sigma=l/(4.*n))
blobs = im > im.mean()
plt.imshow(blobs,cmap="gray")
```

```python
x,y = np.indices((80,80))
x1, y1, x2,y2 = 28, 28,44,52
r1,r2 = 16, 25
mask_circle1 = (x-x1) ** 2 + (y - y1) ** 2 < r1 **2
mask_circle2 = (x-x2) ** 2 + (y - y2) ** 2 < r2 **2
image = np.logical_or(mask_circle1,mask_circle2)
plt.imshow(image)
```

```python
from skimage.morphology import watershed
```

```python
from scipy import ndimage
distance = ndimage.distance_transform_edt(image)
plt.imshow(distance)
```

```python
from skimage.feature import peak_local_max
localmax1 = peak_local_max(distance, indices=False, footprint=np.ones((3,3)), labels=image)
markers = morphology.label(localmax1)
labels = watershed(-distance,markers,mask=image)
plt.imshow(labels)
```

```python
from skimage  import measure
props = measure.regionprops(labels)
[p.area for p in props]
```

```python
[k.perimeter for k in props]
```

```python
mask = coins > filters.threshold_otsu(coins)
plt.imshow(mask)
```

```python
from skimage import segmentation 
clean_border = segmentation.clear_border(mask)
plt.imshow(clean_border)
```

```python
plt.figure()
plt.imshow(coins)
plt.contour(clean_border,cmap = "gray")
```

```python
edges = segmentation.mark_boundaries(coins, clean_border.astype(np.int))
plt.imshow(edges)
```

```python
from skimage.feature import corner_harris, corner_subpix, corner_peaks
from skimage.transform import warp, AffineTransform
```

```python
tform = AffineTransform(scale=(1.3,1.1), rotation=1, shear=0.7, translation=(210,59))
img = warp(data.checkerboard(), tform.inverse, output_shape=(350,350))
plt.imshow(img)
```

```python
peaks = corner_peaks(corner_harris(img), min_distance=5)
plt.imshow(img)
plt.plot(peaks[:,1], peaks[:,0], '.r', markersize=10)
```
